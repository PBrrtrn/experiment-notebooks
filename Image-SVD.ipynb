```python
import matplotlib.pyplot as plt
import numpy as np
import time

from PIL import Image
```

```python
img = Image.open('tmp/vinland_saga.jpg')
plt.figure(figsize=(9,6))
plt.imshow(img)
```

```python
image_matrix = np.array(list(img.getdata(band=0)), float)
image_matrix.shape = (img.size[1], img.size[0])
image_matrix = np.matrix(image_matrix)

plt.figure(figsize=(9,6))
plt.imshow(image_matrix)
```

```python
image_matrix
```

```python
U, sigma, Vt = np.linalg.svd(image_matrix)
```

```python
total_energy = sum(sigma**2)
plt.plot(sigma)
```

```python
plt.plot(sigma[:15])
```

```python
total_energy = sum(sigma**2)
singular_value_energy = sigma**2
energy_percentages = singular_value_energy*100/total_energy

plt.plot(energy_percentages)
```

```python
plt.plot(energy_percentages[:10])
```

```python
recovered_image = np.matrix(U[:,:1]) * np.diag(sigma[:1]) * np.matrix(Vt[:1,:])
plt.figure(figsize=(9,6))
plt.imshow(recovered_image)
```

```python
for i in range(2, 52, 5):
    recovered_image = np.matrix(U[:,:i]) * np.diag(sigma[:i]) * np.matrix(Vt[:i,:])
    plt.figure(figsize=(9,6))
    plt.imshow(recovered_image)
    
```
